```python
import pandas as pd
import collections
import copy
import scipy.sparse as sp
import numpy as np
```

```python
path = 'datasets/ml1m/u.data'
df = pd.read_csv(path, sep='\t', header=None,
        names=['user_id', 'item_id', 'rating', 'time'], index_col=False)
df
```

```python
df_ratehigh = df[df['rating']<=2]
df_ratehigh = df_ratehigh.drop(['time'], axis=1)
df_ratehigh
```

```python
msk = int(len(df_ratehigh) * 0.8)
train = df_ratehigh[:msk]
test = df_ratehigh[msk:]
train.head()
```

```python
train.to_csv('datasets/mlmil/train_under2.csv', index=False, header=False)
test.to_csv('datasets/mlmil/test_under2.csv', index=False, header=False)
```
